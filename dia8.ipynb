# Librerías: pandas & plotly

```python
# Librería para procesar datos
import pandas as pd           
# Librería para crear gráficas
import plotly.express as px   
```

```python
# Crear un DataFrame con los datos de Canada
df = px.data.gapminder().query("country=='Canada'")
```

```python
# Analizamos el contenido
df
```

```python
# Crear la gráfica (con una línea de código)
fig = px.line(df, x="year", y="lifeExp", title="Evolucion de la esperanza de vida")
fig.show()
```

## Comparación de esperanza de vida

```python
# Creamos un DataFrame del continente Oceania
df = px.data.gapminder().query("continent=='Oceania'")
```

```python
# Vemos el contenido del DataFrame
df
```

```python
# Armamos el gráfico
fig2 = px.line(df, x='year', y='lifeExp', color='country')
fig2.show()
```
